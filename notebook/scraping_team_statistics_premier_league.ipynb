## Standard Stats for Serie A Teams

Bu bölüm, web sitesinde Serie A takımlarının standart istatistiklerini çekmekteyiz.

```python
standing_url = "https://fbref.com/en/comps/9/2022-2023/2022-2023-Premier-League-Stats"
data = requests.get(standing_url)
soup = BeautifulSoup(data.text)
```

```python
standings_table = soup.select('table.stats_table')[0]
```

```python
table = soup.find('table', {'id': 'stats_squads_standard_for'})

# Initialize lists to store data
data = []

# Find table rows
rows = table.find_all('tr')

# Skip the header row
for row in rows[1:]:
    columns = row.find_all(['th', 'td'])
    squad = columns[0].text.strip()
    average_age = columns[2].text.strip()
    possession = columns[3].text.strip()
    goals = columns[8].text.strip()
    assists = columns[9].text.strip()
    goal_asist = columns[10].text.strip()
    goalmpk = columns[11].text.strip()
    pk = columns[12].text.strip()
    pk_att = columns[13].text.strip()
    xg = columns[16].text.strip()
    npxg = columns[17].text.strip()
    xag = columns[18].text.strip()
    npxg_xag = columns[19].text.strip()
    prgc = columns[20].text.strip()
    prgp = columns[21].text.strip()
    gls_90 = columns[22].text.strip()
    ast_90 = columns[23].text.strip()
    gls_ast_90 = columns[24].text.strip()
    glsmpk_90 = columns[25].text.strip()
    xg_90 = columns[27].text.strip()
    xag_90 = columns[28].text.strip()
    xg_xag_90 = columns[29].text.strip()
    npxg_90 = columns[30].text.strip()
    npxg_xag_90 = columns[31].text.strip()
    data.append((squad, average_age, possession, goals, assists,goal_asist,goalmpk,pk,pk_att,xg,
                 npxg,xag,npxg_xag,prgc,prgp,gls_90,ast_90,gls_ast_90,glsmpk_90,xg_90,xag_90,xg_xag_90,
                npxg_90,npxg_xag_90))
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
first_row =df.iloc[0]
df.columns = first_row
df.drop(0, inplace=True)
df
```

```python
df_squad_standard = df.copy()
```

```python
df_squad_standard.to_csv(r'D:\Dataset\premier-league\team\squad_premier_league_stats.csv')
```

## Shooting Stats for Premier League

Bu bölümde, takımların şut istatistiklerini çekmekteyiz.
Burada, analiz ettiğiniz şut istatistiklerinin türleri (örneğin, şut sayısı, isabetli şut oranı vb.) ve bu istatistiklerin takım performansını nasıl etkilediği hakkında bilgi alabiliriz.

```python
def scraping_soup(url):
    standing_url = url
    data = requests.get(standing_url)
    soup = BeautifulSoup(data.text)
    standings_table = soup.select('table.stats_table')[0]
    return soup, standings_table
```

```python
soup,standings_table = scraping_soup("https://fbref.com/en/comps/9/2022-2023/shooting/2022-2023-Premier-League-Stats")
```

```python
table = soup.find('table', {'id': 'stats_squads_shooting_for'})

# Initialize lists to store data
data = []

# Find table rows
rows = table.find_all('tr')

# Skip the header row
for row in rows[1:]:
    columns = row.find_all(['th', 'td'])
    squad = columns[0].text.strip()
    goals = columns[3].text.strip()
    shots = columns[4].text.strip()
    shont = columns[5].text.strip()
    shrate = columns[6].text.strip()
    sh90 = columns[7].text.strip()
    shont90 = columns[8].text.strip()
    gdsh = columns[9].text.strip()
    gdshont = columns[10].text.strip()
    pk = columns[13].text.strip()
    xg = columns[15].text.strip()
    npxg = columns[16].text.strip()
    npxgdsh = columns[17].text.strip()
    gmxg = columns[18].text.strip()
    ngmnxg = columns[19].text.strip()
   
    data.append((squad, goals, shots, shont, shrate,sh90,shont90,gdsh,gdshont,pk,xg,
                 npxg,npxgdsh,gmxg,ngmnxg))
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
first_row =df.iloc[0]
df.columns = first_row
df.drop(0, inplace=True)
df
```

```python
df_shoot = df.copy()
```

```python
df.to_csv(r'D:\Dataset\premier-league\team\shooting_premier_league_stats.csv')
```
